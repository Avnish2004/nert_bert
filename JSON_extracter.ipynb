```python
#Extracting id,entity,label and span from list
def extract_entities(words):
    words_List = []
    for x in range(min(6, len(words))):  # Use min to ensure we don't go out of range
        words_List.append(words[x])
    return words_List
```

```python
import re


def clean_and_fix_json(data_string):
    """
    Clean the JSON string by ensuring it ends with a complete entry and proper JSON structure.
    """
    truncated_string = data_string

    # Check if the string ends with a single quote
    if data_string[-1] != "'":
        # Find the position of the last comma
        last_comma_index = data_string.rfind(',')

        if last_comma_index != -1:
            # Truncate the string to include only up to the last comma
            truncated_string = data_string[:last_comma_index ]
        else:
            # If no comma is found, use the original string
            truncated_string = data_string

        # Count and fix bracket imbalance
        open_brackets = truncated_string.count('[')
        close_brackets = truncated_string.count(']')
        if open_brackets > close_brackets:
            # Add missing closing brackets
            truncated_string += ']' * (open_brackets - close_brackets)

        # Ensure the string ends with a single quote if needed
        if not truncated_string.endswith("'"):
            truncated_string += "'"


    return truncated_string


```

```python
#Testing whether our clean function runs properly
row1 = df.iloc[22]['entities']

clean_and_fix_json(row1)
```

```python
import ast
import json

def convert_row_to_json(row):
    text = row['Discharge Summary']
    entities = row['entities']  # This should be a list of dicts

    # Given string (without the trailing ' character for easier parsing)
    string=clean_and_fix_json(entities)

# Convert the string to a Python list using ast.literal_eval
    parsed_list = ast.literal_eval(string)
# Given bytes object
    byte_string = parsed_list

# Decode the bytes object to a string
    decoded_string = byte_string.decode('utf-8')

# Convert the decoded string to a Python list using ast.literal_eval

    data_str = decoded_string
    data_list = json.loads(data_str)



    entity_list = []
    for entity in data_list:

        # Assuming each entity is a dict with keys like 'start', 'end', 'label'
     final_list=extract_entities(entity)
     if len(final_list) >= 6:
        entity_dict = {
        "id": final_list[0],
        "entity": final_list[1],
        "label": final_list[2],
        "start": final_list[4],  # Start position in the text
        "end": final_list[5],    # End position in the text
        }
        entity_list.append(entity_dict)


    return {
        "text": text,
        "entities": entity_list
    }

# Apply the conversion to each row
json_data = df.apply(convert_row_to_json, axis=1).tolist()

#Save the json data
with open('data.json', 'w') as file:
    json.dump(json_data, file, indent=4)
```
